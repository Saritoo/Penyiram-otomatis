#Simulasi Fuzzy Logic

## import library

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

## Variabel input dan output

```python
# Variabel Input
suhu = ctrl.Antecedent(np.arange(0, 101, 1), 'suhu')
kelembaban = ctrl.Antecedent(np.arange(0, 101, 1), 'kelembaban')
lembabUdara = ctrl.Antecedent(np.arange(0, 101, 1), 'lembabUdara')

# Variabel Output
kecepatan_motor = ctrl.Consequent(np.arange(0, 101, 1), 'kecepatan_motor')
```

## Fuzzy Set

```python
# FuzzySet untuk Suhu
suhu['dingin'] = fuzz.trimf(suhu.universe, [0, 0, 15])
suhu['sedang'] = fuzz.trimf(suhu.universe, [10, 25, 40])
suhu['panas'] = fuzz.trimf(suhu.universe, [30, 60, 100])

# FuzzySet untuk Kelembaban
kelembaban['kering'] = fuzz.trimf(kelembaban.universe, [0, 0, 40])
kelembaban['lembap'] = fuzz.trimf(kelembaban.universe, [30, 65, 90])
kelembaban['basah'] = fuzz.trimf(kelembaban.universe, [70, 100, 100])

# FuzzySet untuk lembabUdara
lembabUdara['rendah'] = fuzz.trimf(lembabUdara.universe, [0, 0, 30])
lembabUdara['biasa'] = fuzz.trimf(lembabUdara.universe, [20, 50, 80])
lembabUdara['tinggi'] = fuzz.trimf(lembabUdara.universe, [70, 100, 100])

# FuzzySet untuk Kecepatan Motor
kecepatan_motor['sangat_cepat'] = fuzz.trimf(kecepatan_motor.universe, [0, 0, 20])
kecepatan_motor['cepat'] = fuzz.trimf(kecepatan_motor.universe, [10, 25, 40])
kecepatan_motor['normal'] = fuzz.trimf(kecepatan_motor.universe, [30, 40, 50])
kecepatan_motor['lambat'] = fuzz.trimf(kecepatan_motor.universe, [30, 50, 70])
kecepatan_motor['mati'] = fuzz.trimf(kecepatan_motor.universe, [60, 100, 100])
```

## Fuzzy Rule

```python
# Aturan Fuzzy
rule1 = ctrl.Rule(suhu['dingin'] & kelembaban['kering'] & lembabUdara['rendah'], kecepatan_motor['cepat'])
rule2 = ctrl.Rule(suhu['dingin'] & kelembaban['kering'] & lembabUdara['biasa'], kecepatan_motor['cepat'])
rule3 = ctrl.Rule(suhu['dingin'] & kelembaban['kering'] & lembabUdara['tinggi'], kecepatan_motor['cepat'])
rule4 = ctrl.Rule(suhu['dingin'] & kelembaban['lembap'] & lembabUdara['rendah'], kecepatan_motor['sangat_cepat'])
rule5 = ctrl.Rule(suhu['dingin'] & kelembaban['lembap'] & lembabUdara['biasa'], kecepatan_motor['sangat_cepat'])
rule6 = ctrl.Rule(suhu['dingin'] & kelembaban['lembap'] & lembabUdara['tinggi'], kecepatan_motor['sangat_cepat'])
rule7 = ctrl.Rule(suhu['dingin'] & kelembaban['basah'] & lembabUdara['rendah'], kecepatan_motor['sangat_cepat'])
rule8 = ctrl.Rule(suhu['dingin'] & kelembaban['basah'] & lembabUdara['biasa'], kecepatan_motor['sangat_cepat'])
rule9 = ctrl.Rule(suhu['dingin'] & kelembaban['basah'] & lembabUdara['tinggi'], kecepatan_motor['sangat_cepat'])
rule10 = ctrl.Rule(suhu['sedang'] & kelembaban['kering'] & lembabUdara['rendah'], kecepatan_motor['lambat'])
rule11 = ctrl.Rule(suhu['sedang'] & kelembaban['kering'] & lembabUdara['biasa'], kecepatan_motor['lambat'])
rule12 = ctrl.Rule(suhu['sedang'] & kelembaban['kering'] & lembabUdara['tinggi'], kecepatan_motor['lambat'])
rule13 = ctrl.Rule(suhu['sedang'] & kelembaban['lembap'] & lembabUdara['rendah'], kecepatan_motor['cepat'])
rule14 = ctrl.Rule(suhu['sedang'] & kelembaban['lembap'] & lembabUdara['biasa'], kecepatan_motor['cepat'])
rule15 = ctrl.Rule(suhu['sedang'] & kelembaban['lembap'] & lembabUdara['tinggi'], kecepatan_motor['cepat'])
rule16 = ctrl.Rule(suhu['sedang'] & kelembaban['basah'] & lembabUdara['rendah'], kecepatan_motor['sangat_cepat'])
rule17 = ctrl.Rule(suhu['sedang'] & kelembaban['basah'] & lembabUdara['biasa'], kecepatan_motor['sangat_cepat'])
rule18 = ctrl.Rule(suhu['sedang'] & kelembaban['basah'] & lembabUdara['tinggi'], kecepatan_motor['sangat_cepat'])
rule19 = ctrl.Rule(suhu['panas'] & kelembaban['kering'] & lembabUdara['rendah'], kecepatan_motor['mati'])
rule20 = ctrl.Rule(suhu['panas'] & kelembaban['kering'] & lembabUdara['biasa'], kecepatan_motor['mati'])
rule21 = ctrl.Rule(suhu['panas'] & kelembaban['kering'] & lembabUdara['tinggi'], kecepatan_motor['mati'])
rule22 = ctrl.Rule(suhu['panas'] & kelembaban['lembap'] & lembabUdara['rendah'], kecepatan_motor['lambat'])
rule23 = ctrl.Rule(suhu['panas'] & kelembaban['lembap'] & lembabUdara['biasa'], kecepatan_motor['lambat'])
rule24 = ctrl.Rule(suhu['panas'] & kelembaban['lembap'] & lembabUdara['tinggi'], kecepatan_motor['lambat'])
rule25 = ctrl.Rule(suhu['panas'] & kelembaban['basah'] & lembabUdara['rendah'], kecepatan_motor['cepat'])
rule26 = ctrl.Rule(suhu['panas'] & kelembaban['basah'] & lembabUdara['biasa'], kecepatan_motor['cepat'])
rule27 = ctrl.Rule(suhu['panas'] & kelembaban['basah'] & lembabUdara['tinggi'], kecepatan_motor['cepat'])
```

## Control Fuzzy dan input

```python
# Sistem Kontrol Fuzzy
sistem_kontrol = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9,
                                     rule10, rule11, rule12, rule13, rule14, rule15, rule16, rule17, rule18,
                                     rule19, rule20, rule21, rule22, rule23, rule24, rule25, rule26, rule27])
motor_ctrl = ctrl.ControlSystemSimulation(sistem_kontrol)

# Masukkan nilai fuzzy ke dalam sistem
motor_ctrl.input['suhu'] = 25
motor_ctrl.input['kelembaban'] = 50
motor_ctrl.input['lembabUdara'] = 75

# Lakukan perhitungan fuzzy
motor_ctrl.compute()
```

## Output Fuzzy

### hasil


```python
# Output kecepatan motor
print("Kecepatan Motor:", motor_ctrl.output['kecepatan_motor'])
```

### plot

```python
suhu.view(sim=motor_ctrl)
kelembaban.view(sim=motor_ctrl)
lembabUdara.view(sim=motor_ctrl)
kecepatan_motor.view(sim=motor_ctrl)

# Tampilkan plot
plt.show()
```
